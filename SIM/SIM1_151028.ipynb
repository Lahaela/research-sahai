```python
from __future__ import division
normrnd = random.normal
# from matplotlib import rc
# rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
# rc('text', usetex=True)
```

```python
# System Parameters -- Scalars

a = 1.1 # growth of the system
b = 1 # rate of info, bits / time
D = 3 # delay
alph = 2/3 # rate of channel
code = (D+1)*b # bits in codeword
R = alph * code # bits of information
drop_p = 2^(-(code - R + 1)) # drop probability
```

```python
# Noise Parameters

varx = 1
varw = 1
varv = 1
mu_c = 1
varc = 2^(-2*R)
```

```python
n = 50*D # time of simulation
trials = 1000 # number of trials
XD = zeros((trials, n)) # aggregate info from all trials, (row, col)

ud = a**D*mu_c/(mu_c**2+varc) # control for delay
```

```python
for m in range(trials):
    xd, yd = zeros(n), zeros(n)
    xd[0] = normrnd(0, sqrt(varx))
    #xd = normrnd(0,sqrt(varx),[1,1]) # x[0] ~ N(0, 1)
    c = normrnd(mu_c,sqrt(varc)) # c(0) ~ N(1, sigma_c)
    yd[0] = c*xd[0]+normrnd(0, sqrt(varv))
    #yd = c*xd+normrnd(0,sqrt(varv))
    
    for t in range(n):
        w = normrnd(0,sqrt(varw))
        v = normrnd(0,sqrt(varv))
        c = normrnd(mu_c,sqrt(varc))
        
#         control = ud*yd[t-D]
#         if abs(control) > thresh:
#             control = thresh*control/abs(control) #preserve sign

        if t % (D+1) == 0 and random.random() > drop_p:
            xd[t] = a*(xd[t-1]-ud*yd[t-D])+w
            #xd = [xd a*(xd(t)-ud*yd(t-D))+w]
        else:
            #xd = [xd a*xd(t)+w]
            xd[t] = a*xd[t-1]+w

        yd[t] = c*xd[t]+v
        #yd = [yd c*xd(t+1)+v]
    
#     XD = [XD; xd];
    XD[m] = xd
```

```python
# Plot of Absolute Value of X

plot(mean(absolute(XD), axis=0), lw=2.0) # mean along vertical axis
title('Mean(Abs(X))')
xlabel('n = time')
```

```python
# Plot of Variance of X

ZD = [x**2 for x in XD]
plot(mean(ZD, axis=0), lw=2.0)
title('Mean($X^2$)')
```
