```python
# from operator import mul    # or mul=lambda x,y:x*y
# from fractions import Fraction

import operator as op
def nCr(n, r):
    r = min(r, n-r)
    if r == 0: return 1
    numer = reduce(op.mul, xrange(n, n-r, -1))
    denom = reduce(op.mul, xrange(1, r+1))
    return numer//denom

# def nCr(n,r): 
#   return int( reduce(mul, (Fraction(n-i, i+1) for i in range(r)), 1) )

# def nCr(n,r):
#     f = math.factorial
#     return f(n) / f(r) / f(n-r)
```

$$ P(bitdrop) = \frac{1}{2} \, \text{erfc}\left(\sqrt{\frac{\text{Operating SNR}}{2}} \right) $$

Probability Hamming syllable fails:
$$ hcerr = 1 - \left( (1-p_d)^7 + {7 \choose 1} p_d(1-p_d)^6 \right)$$

Probability RS codeword fails:
$$ hcf = 1 - \left( (1-p_d)^{14} + {14 \choose 1} p_d(1-p_d)^{13} + \left( {7 \choose 1} p_d(1-p_d)^6 \right)^2 \right) $$
$$ hcf = 1 - (1-hcerr)^2 $$

Probability Reed Solomon message fails:
$$ \text{rsfail } = \sum_{d= \lfloor \frac{k}{2} \rfloor +1}^{20+k} {20+k \choose d} (hcf)^d (1-hcf)^{20+k-d}$$
where d stands for dropped RS words, and this sum represents enough of the codeword is dropped that important words start being dropped as well. All of k (n+k) is dropped, and the 'final' 20 words in the payload (that are necessary for decoding) are now being encroached upon.) k is in the range (0, 236) so that the final blocklength is in the range (20, 256).

```python
# 4,3 Hamming Code - Reed Solomon Wrap
def hs_rs_table(rate, blocklength):
    k = (1-rate)*blocklength
    msg = rate*blocklength
    res = {}
    op_SNR = 1
    while op_SNR < 10**(5): # 10 = 100 dB
        pbitdrop = 0.5*math.erfc(sqrt(op_SNR/2))
        hcerr = 1 - ((1-pbitdrop)**7 + 7*pbitdrop*(1-pbitdrop)**6)
        hcf = 1 - (1-hcerr)**2
        reeddrop = sum([nCr(blocklength, d)* hcf**d *(1-hcf)**(blocklength-d) for d in range(int(k/2)+1, blocklength)])
        res[op_SNR] = reeddrop
        op_SNR *= 10**(0.01)
    return res
    
def print_table_csv(table, filename):
    ks = sorted(table.keys())
    vs = [table[k] for k in ks]
    with open(filename+'.csv', 'wb') as csvfile:
        writer = csv.writer(csvfile, delimiter=',')
        writer.writerow(['SNR (dB)', 'P(fail)'])
        for rowid in xrange(len(ks)):
            writer.writerow([str(log10(ks[rowid])),str(vs[rowid])])
            
test = hs_rs_table(4/7, 256)
print_table_csv(test, 'hs_rs_table')
```

```python
# Intermediate Sandbox for above

rate = 4/7
blocklength = 256
k = (1-rate)*blocklength
msg = rate*blocklength
res = {}
op_SNR = 1
while op_SNR < 10**(1): # 10 = 100 dB
    pbitdrop = 0.5*math.erfc(sqrt(op_SNR/2))
    hcerr = 1 - ((1-pbitdrop)**7 + 7*pbitdrop*(1-pbitdrop)**6)
    hcf = 1 - (1-hcerr)**2
    reeddrop = sum([nCr(blocklength, d)* hcf**d *(1-hcf)**(blocklength-d) for d in range(int(k/2)+1, blocklength)])
    res[op_SNR] = reeddrop
    print('SNR', op_SNR, 'Pbitdrop', pbitdrop, 'P(fail)', reeddrop)
    op_SNR *= 10**(0.01)
```

```python
# Assumes table is in order, does linear interpolation
# dSNR = 0.01 dB for 4,3 Hamming Lookup Table
# op_SNR in dB
def p_snr_lookup_fail(filename, dSNR, op_SNR):
    if op_SNR < 0.0: # this might change based on coding schemes
        return 1.0
    if op_SNR > 5.0:
        return 0.0
    with open(filename, 'rb') as csvfile:
        reader = csv.reader(csvfile, delimiter=',')
        endpts = []
        for row in reader:
#             if row[0] != 'SNR (dB)':
#                 snr = float(row[0])
            if row[0] == 'SNR (dB)':
                continue
            snr = float(row[0])
            if snr == op_SNR: # exact lookup
                return float(row[1])
            elif (op_SNR - dSNR) < snr < (op_SNR+dSNR):
                endpts.append((snr, float(row[1])))
            elif snr > (op_SNR + dSNR):
                break
    slope = endpts[1][1] - endpts[0][1]
    dx = op_SNR - endpts[0][0]
    return slope*dx/dSNR + endpts[0][1]
            
res = p_snr_lookup_fail('hs_rs_table.csv', 0.01, 0.383)
print(res)
```

```python
# Assumes op_SNR is in table
def p_snr_lookup_fail_2(filename, op_SNR):
        with open(filename, 'rb') as csvfile:
        reader = csv.reader(csvfile, delimiter=',')
        for row in reader:
            if row[0] == 'SNR':
                continue
            snr = float(row[0])
            if snr == op_SNR: # exact lookup
                return float(row[1])
```

```python
# Codetable = 'filename.csv' for chosen coding scheme
def p_single(codetable, op_SNR, endpoint):
    fadexp = sp.stats.expon() # scale = 1.0 = 1/lambda
    # 0.01 = dSNR for 4,3 Hamming Code table
    def fadefunc(fade):
        eff_SNR = log10(op_SNR*fade)
        pfail = p_snr_lookup_fail(codetable, 0.01, eff_SNR)
        return pfail * fadexp.pdf(fade)
    
    return sp.integrate.quad(fadefunc, 0, endpoint)
    
# ffade = sp.stats.expon(scale=1.0)
# y = p_single('hs_rs_table.csv', 1, ffade, 50)
```

```python
# Integrating fade 0 to 50
x = arange(0, log10(10**5/50), 0.01)
SNR = 10**x
y = zip(*[p_single('hs_rs_table.csv', op_snr, 50) for op_snr in SNR])[0]
# z = zip(*y) throwing away the error tolerance bounds (should be < 10**(-6)
plot(10*x, y, lw=2.0)
title('P(single fail)', fontsize=24)
xlabel('Operating SNR (dB)', fontsize=20)
```

```python
def p_combo(codetable, a, op_SNR, endpoint):
    fadexp = sp.stats.erlang(a)
    # 0.01 = dSNR for 4,3 Hamming Code table
    def fadefunc(fade):
        eff_SNR = log10(op_SNR*fade)
        pfail = p_snr_lookup_fail(codetable, 0.01, eff_SNR)
        return pfail * fadexp.pdf(fade)
    
    return sp.integrate.quad(fadefunc, 0, endpoint)
```

```python
# Integrating fade 0 to 50, a = 1
x = arange(0, log10(10**5/50), 0.01)
SNR = 10**x
y = zip(*[p_combo('hs_rs_table.csv', 1, op_snr, 50) for op_snr in SNR])[0]
# z = zip(*y) throwing away the error tolerance bounds (should be < 10**(-6)
plot(10*x, y, lw=2.0)
title('P(combo fail)', fontsize=24)
xlabel('Operating SNR (dB)', fontsize=20)
```

```python
# N = num_nodes
def p_protocol(codetable, N, op_SNR, endpoint):
    psingle = p_single(codetable, op_SNR, endpoint)[0]
    return sum([nCr(N, a) * (1-psingle)**a * psingle**(N-a) * (1-(1-p_combo(codetable, a, op_SNR, endpoint)[0])**(N-a)) for a in range(N)])
```

```python
# N = 10
x = arange(0, log10(10**5/50), 0.01)
SNR = 10**x
y = [p_protocol('hs_rs_table.csv', 10, op_snr, 50) for op_snr in SNR]
plot(10*x, y, lw=2.0)
title('P(Protocol fail)', fontsize=24)
xlabel('Operating SNR (dB)', fontsize=20)
```
