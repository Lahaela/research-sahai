## Finding the Nominal SNR Needed

General Errors

$$ \text{Number of nodes} = N $$
$$ \text{Rate} = \frac{N \cdot \text{msg_size} }{\text{blocklength}} $$

Individual Hamming code error:
$$ hcerr = 1 - \left( (1-p_{bit})^7 + {7 \choose 1} \, p_{bit} \, (1-p_{bit})^6 \right)$$

2 words together: Hamming code error:
$$ \text{P(Hamming code fails)} = hcf = 1 - \left( (1-p_{bit})^{14} + {14 \choose 1} \, p_{bit} \, (1-p_{bit})^{13} + \left( {7 \choose 1} \, p_{bit} \, (1-p_{bit})^6 \right)^2 \right) = 1 - (1-hcerr)^2 $$

3 words together:
$$ hcf = 1 - (1-hcerr)^3 $$

Blocklength: $\lfloor \frac{10,000 \text{ bits}}{14 \text{ bits per codeword}} \rfloor = 714$
$$ \text{P(Reed Solomon Code fails)} = \sum_{d=\lfloor \frac{k}{2} \rfloor + 1}^{\text{block}} {\text{block} \choose d} \, hcf^d \, (1-hcf)^{\text{block}-d} = \text{P(Additive Noise)}$$
$$ \text{P(Additive Noise)} < 10^{-9} \rightarrow p_{crit} $$
$$ \text{Actual SNR} = \left( Q^{-1}(p_{crit}) \right)^2 = \left( \sqrt{2} \, \text{erf}^{-1} (1 - 2p_{crit}) \right)^2 $$

$$ \text{P(protocol)} = \sum_{a = 0}^{N-1} {N \choose a} \, (1-p_{nominal})^a \, p_{nominal}^{N-a} \, \left( 1-(1-p_{badfade}^a - p_{add})^{N-a} \right) $$
$$ \text{P(protocol)} = 10^{-9} \rightarrow p_{nominal} $$

$$ p_{nominal} = \text{P(bad fade)} + \text{P(Additive Noise)} $$

$$ \text{P(bad fade)} = 1 - \text{exp} \left( \frac{-\text{Actual SNR}}{\text{Nominal SNR}} \right) $$

$$ \text{Nominal SNR} = \frac{-\text{Actual SNR}}{ln (1 - (p_{nominal} - p_{add}))} $$

```python
numwords = 3 # number of hamming code blocks stuck together
bpw = 21 # bits per word
detect = 2 # if general errors --> detect = 2, erasures --> detect = 1

param = 10**(-9)
msgsize = 160
blocklength = 10000
rsblock = int(blocklength/bpw)

nomSNR = []
padd = []
nomtracker = []
actual = []
for num_nodes in range(1, 36): # [1, 35]
    rate = num_nodes*msgsize/blocklength
    k = (1-rate)*rsblock
    
    pdbit = arange(0,0.15,0.001) # bit error prob
    hcerr = 1 - ((1-pdbit)**7 + (detect-1)*7*pdbit*(1-pdbit)**6)
#     hammfail = 1 - ((1-pdbit)**14 + 14*pdbit*(1-pdbit)**13 + (7*pdbit*(1-pdbit)**6)**2)
    hammfail = 1 - (1-hcerr)**numwords

    for hcf in hammfail:
        reed = sum([nCr(rsblock, d)* hcf**d *(1-hcf)**(rsblock-d) for d in range(int(k/detect)+1, rsblock)])
        if reed > param:
            break
    idx = np.where(hammfail==hcf)[0][0]
    hccrit = hammfail[idx-1]
    pbcrit = pdbit[idx-1]
    minSNR = (sqrt(2)*sp.special.erfinv(1-2*pbcrit))**2 #given bit drop prob min SNR needed to decode, 'Actual SNR'
    actual.append(minSNR)
    paddnoise = sum([nCr(rsblock, d)* hccrit**d *(1-hccrit)**(rsblock-d) for d in range(int(k/detect)+1, rsblock)])

    nominalSNR = 10**0
    pprotocol = 1
    while pprotocol > 10**(-9):
        nominalSNR *= 10**(0.01)
        pbadfade = 1 - math.exp(-minSNR/nominalSNR)
        pnom = pbadfade + paddnoise # p nominal
        pprotocol = sum([nCr(num_nodes, a) * (1-pnom)**a * pnom**(num_nodes-a) * (1-(1-pbadfade**a-paddnoise)**(num_nodes-a)) for a in range(num_nodes)])
    
#     pdb2 = arange(0, 0.3, 0.001)
#     plink = [sum([nCr(num_nodes, a) * (1-pdb)**a * pdb**(num_nodes-a) * (1-(1-pdb**a)**(num_nodes-a)) for a in range(num_nodes)]) for pdb in pdb2]
#     idfade = np.where(array(plink)<=param)[0][-1]
#     nominal_pb = pdb2[idfade]
#     nomtracker.append(nominal_pb)
    
#     nominalSNR = (sqrt(2)*sp.special.erfinv(1-2*nominal_pb))**2
#     pbadfade = 1 - math.exp(-minSNR/nominalSNR)
#     paddnoise = sum([nCr(rsblock, d)* hccrit**d *(1-hccrit)**(rsblock-d) for d in range(int(k/2)+1, rsblock)]) # pr reed solomon drop
    padd.append(paddnoise)
#     nominalSNR = -minSNR / log(1-nominal_pb+paddnoise)
    nomSNR.append(nominalSNR)
    nomtracker.append(pnom)
#     plink.append(pbadfade+paddnoise)
```

```python
# Erasures / Detection Mode

param = 10**(-9)
msgsize = 160
blocklength = 10000
rsblock = int(blocklength/14)

nomSNR = []
padd = []
nomtracker = []
for num_nodes in range(1, 36): # [1, 35]
    rate = num_nodes*msgsize/blocklength
    k = (1-rate)*rsblock
    
    pdbit = arange(0,0.15,0.001) # bit error prob
    hammfail = 1 - ((1-pdbit)**14)

    for hcf in hammfail:
        reed = sum([nCr(rsblock, d)* hcf**d *(1-hcf)**(rsblock-d) for d in range(int(k)+1, rsblock)])
        if reed > param:
            break
    idx = np.where(hammfail==hcf)[0][0]
    hccrit = hammfail[idx-1]
    pbcrit = pdbit[idx-1]
    minSNR = (sqrt(2)*sp.special.erfinv(1-2*pbcrit))**2 #given bit drop prob min SNR needed to decode, 'Actual SNR'
    paddnoise = sum([nCr(rsblock, d)* hccrit**d *(1-hccrit)**(rsblock-d) for d in range(int(k)+1, rsblock)])

    nominalSNR = 10**0
    pprotocol = 1
    while pprotocol > 10**(-9):
        nominalSNR *= 10**(0.01)
        pbadfade = 1 - math.exp(-minSNR/nominalSNR)
        pnom = pbadfade + paddnoise # p nominal
        pprotocol = sum([nCr(num_nodes, a) * (1-pnom)**a * pnom**(num_nodes-a) * (1-(1-pbadfade**a-paddnoise)**(num_nodes-a)) for a in range(num_nodes)])
    
    padd.append(paddnoise)
    nomSNR.append(nominalSNR)
    nomtracker.append(pnom)
```

# Fixed Payload, Max Tolerable Bit Drop Probability

## 3-Repetition Code

**Needs Updating**

Payload = 160 bits $\rightarrow$ 160 repetition code blocks = 20 RS words

Rates in range $[\frac{5}{192}, \frac{1}{3}] \rightarrow [\frac{160}{6144}, \frac{160}{480}]$, with the denominator increasing by 24 bits (1 RS word.)

Repetition code = 1 information bit, 2 parity bits, max rate = 1/3

8 Repetition codewords put together $\rightarrow$ 8 information bits, 16 parity bits, total 24 bits, max rate = 1/3

Reed-Solomon in field $2^8 = 256$ words, $256 \cdot 24 = 6,144$ bits

Probability repetition codeword fails:
$$ rcf = 1 - \left( (1-p_d)^{3} + {3 \choose 1} p_d(1-p_d)^{2} \right) $$
$$ \text{8-fail} = 1 - (1-rcf)^8 $$

Probability Reed Solomon message fails:
$$ \text{rsfail } = \sum_{d= \lfloor \frac{k}{2} \rfloor +1}^{20+k} {20+k \choose d} (hcf)^d (1-hcf)^{20+k-d}$$
where d stands for dropped RS words, and this sum represents enough of the codeword is dropped that important words start being dropped as well. All of k (n+k) is dropped, and the 'final' 20 words in the payload (that are necessary for decoding) are now being encroached upon.) k is in the range (0, 236) so that the final blocklength is in the range (20, 256).

$$ \text{rsfail } < 10^{-5} $$

```python
# Fixed payload comparison
comparison = []
param = 10**(-5) # overall tolerable error
```

```python
# 3-repetition code
numsyl = 8

maxpd = []
pdbit = arange(0,0.5,0.001) # bit error prob
hcerr = 1 - ((1-pdbit)**3 + 3*pdbit*(1-pdbit)**2)
hammfail = 1 - (1-hcerr)**numsyl

rates = [160/(block*(3*numsyl)) for block in xrange(20,257)]

for k in xrange(0, 237):
    reeddrop = [sum([nCr(20+k, d)* hcf**d *(1-hcf)**(20+k-d) for d in range(int(k/2)+1, 20+k)]) for hcf in hammfail]
    reedfail = filter(lambda x: x < param, reeddrop)
    maxpd.append(pdbit[(len(reedfail)-1)])
    
# figure()
stem(rates, maxpd, lw=2.0)
xlabel('Rate', fontsize=20)
ylabel('Max Bit Error Probability', fontsize=18, labelpad=15)
title('Overall Tolerable Drop Pr = {0}'.format(param), fontsize=24)

comparison.append((rates, maxpd))
```

## 5-Repetition Code

**Needs Updating**

Repetition code = 1 information bit, 4 parity bits, max rate = 1/5

8 Repetition codewords put together $\rightarrow$ 8 information bits, 32 parity bits, total 40 bits, max rate = 1/5

Payload = 160 bits $\rightarrow$ 160 repetition code blocks = 20 RS words

Rates in range $[\frac{1}{64}, \frac{1}{5}] \rightarrow [\frac{160}{10240}, \frac{160}{800}]$, with the denominator increasing by 40 bits (1 RS word.)

Reed-Solomon in field $2^8 = 256$ words, $256 \cdot 40 = 10,240$ bits

Probability repetition codeword fails:
$$ rcf = 1 - \left( (1-p_d)^{5} + {5 \choose 1} p_d(1-p_d)^{4} + {5 \choose 2} p_d^2 (1-p_d)^3 \right) $$
$$ \text{8-fail} = 1 - (1-rcf)^8 $$

Probability Reed Solomon message fails:
$$ \text{rsfail } = \sum_{d= \lfloor \frac{k}{2} \rfloor +1}^{20+k} {20+k \choose d} (hcf)^d (1-hcf)^{20+k-d}$$
where d stands for dropped RS words, and this sum represents enough of the codeword is dropped that important words start being dropped as well. All of k (n+k) is dropped, and the 'final' 20 words in the payload (that are necessary for decoding) are now being encroached upon.) k is in the range (0, 236) so that the final blocklength is in the range (20, 256).

$$ \text{rsfail } < 10^{-5} $$

```python
# 5-repetition code
numsyl = 8

maxpd = []
pdbit = arange(0,0.5,0.001) # bit error prob
hcerr = 1 - ((1-pdbit)**5 + 5*pdbit*(1-pdbit)**4 + 10*pdbit**2*(1-pdbit)**3)
hammfail = 1 - (1-hcerr)**numsyl

rates = [160/(block*(5*numsyl)) for block in xrange(20,257)]

for k in xrange(0, 237):
    reeddrop = [sum([nCr(20+k, d)* hcf**d *(1-hcf)**(20+k-d) for d in range(int(k/2)+1, 20+k)]) for hcf in hammfail]
    reedfail = filter(lambda x: x < param, reeddrop)
    maxpd.append(pdbit[(len(reedfail)-1)])
    
# figure()
stem(rates, maxpd, lw=2.0)
xlabel('Rate', fontsize=20)
ylabel('Max Bit Error Probability', fontsize=18, labelpad=15)
title('Overall Tolerable Drop Pr = {0}'.format(param), fontsize=24)
# savefig('160225_rep5_pbiterror.pdf', bbox='tight')

comparison.append((rates, maxpd))
```

## 4, 3 Hamming Code

160 information bits = 40 Hamming syllables = 20 Reed-Solomon words

Rates in range $[\frac{5}{112}, \frac{4}{7}] \rightarrow [\frac{160}{3584}, \frac{160}{280}]$, with the denominator increasing by 14 bits (1 RS word.)

Hamming code = 4 information bits, 3 parity bits, max rate = 4/7

Two Hamming codewords put together $\rightarrow$ 8 information bits, 6 parity bits, max rate = 4/7

Reed-Solomon in field $2^8 = 256$, $256 \cdot 14 = 3,584$

Probability Hamming syllable fails:
$$ hcerr = 1 - \left( (1-p_d)^7 + {7 \choose 1} p_d(1-p_d)^6 \right)$$

Probability RS codeword fails:
$$ hcf = 1 - \left( (1-p_d)^{14} + {14 \choose 1} p_d(1-p_d)^{13} + \left( {7 \choose 1} p_d(1-p_d)^6 \right)^2 \right) $$
$$ hcf = 1 - (1-hcerr)^2 $$

Probability Reed Solomon message fails:
$$ \text{rsfail } = \sum_{d= \lfloor \frac{k}{2} \rfloor +1}^{20+k} {20+k \choose d} (hcf)^d (1-hcf)^{20+k-d}$$
where d stands for dropped RS words, and this sum represents enough of the codeword is dropped that important words start being dropped as well. All of k (n+k) is dropped, and the 'final' 20 words in the payload (that are necessary for decoding) are now being encroached upon.) k is in the range (0, 236) so that the final blocklength is in the range (20, 256).

$$ \text{rsfail } < 10^{-5} $$

```python
# 4,3 Hamming Code

maxpd = []
pd = arange(0,0.5,0.001) # bit error prob
hcerr = 1 - ((1-pd)**7 + 7*pd*(1-pd)**6)
hammfail = 1 - (1-hcerr)**2
# hammfail = 1 - ((1-pd)**14 + 14*(1-pd)**13*pd + (7*(1-pd)**6*pd)**2)

rates = [160/(block*14) for block in xrange(20,257)] # 11 exclusive

for k in xrange(0, 237):
    reeddrop = [sum([nCr(20+k, d)* hcf**d *(1-hcf)**(20+k-d) for d in range(int(k/2)+1, 20+k)]) for hcf in hammfail]
    reedfail = filter(lambda x: x < param, reeddrop)
    maxpd.append(pd[(len(reedfail)-1)])
    
# figure()
stem(rates, maxpd, lw=2.0)
xlabel('Rate', fontsize=20)
ylabel('Max Bit Error Probability', fontsize=18, labelpad=15)
title('Overall Tolerable Drop Pr = {0}'.format(param), fontsize=24)

comparison.append((rates, maxpd))
```

## 7,11 Hamming Code

7,11 Hamming code = 7 information bit, 4 parity bits, max rate = 7/11

2 Repetition codewords put together $\rightarrow$ 14 information bits, 8 parity bits, total 22 bits, max rate = 7/11

Payload = 160 bits $\rightarrow$ = $\approx 11$ RS words = 154 bits

Reed-Solomon in field $2^{14} = 16,384$ words, $16384 \cdot 22 = 360,448$ bits. Since that's huge, I decided to stop at $\approx 10,000$ bits.

Rates in range $[\frac{40}{2497}, \frac{7}{11}] \rightarrow [\frac{154}{454*22}=\frac{154}{9,988}, \frac{154}{242}]$, with the denominator increasing by 22 bits (1 RS word.)

Probability repetition codeword fails:
$$ rcf = 1 - \left( (1-p_d)^{11} + {11 \choose 1} p_d(1-p_d)^{10} \right) $$
$$ \text{2-fail} = 1 - (1-rcf)^2 $$

Probability Reed Solomon message fails:
$$ \text{rsfail } = \sum_{d= k +1}^{11+k} {11+k \choose d} (hcf)^d (1-hcf)^{11+k-d}$$
where d stands for dropped RS words, and this sum represents enough of the codeword is dropped that important words start being dropped as well. All of k (n+k) is dropped, and the 'final' 11 words in the payload (that are necessary for decoding) are now being encroached upon.) k is in the range (0, 443) so that the final blocklength is in the range (11, 454).

$$ \text{rsfail } < 10^{-5} $$

```python
# 7,11 Hamming code -- has detection!
numsyl = 2

maxpd = []
pdbit = arange(0,0.5,0.001) # bit error prob
hcerr = 1 - ((1-pdbit)**(11) + 11*pdbit*(1-pdbit)**(10))
hammfail = 1 - (1-hcerr)**numsyl

rates = [154/(numwords*(11*numsyl)) for numwords in xrange(11,454)]

for k in xrange(0, 443):
    reeddrop = [sum([nCr(11+k, d)* hcf**d *(1-hcf)**(11+k-d) for d in range(int(k)+1, 11+k)]) for hcf in hammfail]
    reedfail = filter(lambda x: x < param, reeddrop)
    maxpd.append(pdbit[(len(reedfail)-1)])
    
# figure()
stem(rates, maxpd, lw=2.0)
xlabel('Rate', fontsize=20)
ylabel('Max Bit Error Probability', fontsize=18, labelpad=15)
title('Overall Tolerable Drop Pr = {0}'.format(param), fontsize=24)
# savefig('160225_hc711_pbiterror.pdf', bbox='tight')

comparison.append((rates, maxpd))
```

## Detection Mode

No Detection: 1-bit flip okay, 2-bit flips $\implies$ corruption

Detection: 1-bit flip / 2-bit flip $\implies$ erasure, 3+ bit flip $\implies$ corruption

160 information bits = 40 Hamming syllables = 20 Reed-Solomon words

Rates in range $[\frac{5}{112}, \frac{4}{7}] \rightarrow [\frac{160}{3584}, \frac{160}{280}]$, with the denominator increasing by 14 bits (1 RS word.)

Hamming code = 4 information bits, 3 parity bits, max rate = 4/7

Two Hamming codewords put together $\rightarrow$ 8 information bits, 6 parity bits, max rate = 4/7

Reed-Solomon in field $2^8 = 256$, $256 \cdot 14 = 3,584$

Probability Hamming syllable fails:
$$ hcdrop = {7 \choose 1} p_d (1-p_d)^6 + {7 \choose 2} p_d^2 (1-p_d)^5 $$
$$ hcflip = 1 - hcdrop - (1-p_d)^7 $$

Probability RS codeword fails:
$$ rsflip = hcflip^2 $$
$$ rsdrop = 1 - (1-p_d)^{14} - hcflip^2 $$

OLD >> Probability Reed Solomon message fails:
$$ \text{rsfail } = \sum_{d= \lfloor \frac{k}{2} \rfloor +1}^{20+k} {20+k \choose d} (hcf)^d (1-hcf)^{20+k-d}$$
where d stands for dropped RS words, and this sum represents enough of the codeword is dropped that important words start being dropped as well. All of k (n+k) is dropped, and the 'final' 20 words in the payload (that are necessary for decoding) are now being encroached upon.) k is in the range (0, 236) so that the final blocklength is in the range (20, 256).

NEW >> Probability Reed Solomon message succeeds:
$$ \sum_{s=\lceil \text{payload + 1/2 k} \rceil}^{\text{block}} {block \choose s} \, succ^s \, (rsdrop + rsflip)^{block - s}  + \sum_{s = \text{payload}}^{\lceil \text{payload + 1/2 k} \rceil} {block \choose s} \, succ^s \left( \sum_{j = \lceil \frac{block - s}{2} \rceil}^{block -s} rsdrop^j rsflip^{block-s-j} \right) $$

$$ \text{rsfail } < 10^{-5} $$

```python
# 4,3 Hamming Code

payload = 20

maxpd = [] 
pd = arange(0,0.5,0.001) # bit error prob
# hcerr = 1 - ((1-pd)**7 + 7*pd*(1-pd)**6)
# hammfail = 1 - (1-hcerr)**2
# hammfail = 1 - ((1-pd)**14 + 14*(1-pd)**13*pd + (7*(1-pd)**6*pd)**2)

rates = [160/(block*14) for block in xrange(payload,257)] # 11 exclusive

for block in xrange(payload, 257): # increasing blocklength, fixed payload
    prfail = []
    for pbd in pd:
        hcdrop = 7*pbd*(1-pbd)**6 + 21*pbd**2*(1-pbd)**5
        hcflip = 1 - hcdrop - (1-pbd)**7
        rsflip = hcflip**2
        rsdrop = 1 - rsflip - (1-pbd)**14
        rs_success = 1 - rsflip - rsdrop
        rsfail = []
        for d in xrange(0, block):
            print('Block', block, 'd', d, block-d-payload, int((block-d-payload)/2+1))
            corrupt = sum([nCr(block-d, j) * rsflip**j * rs_success**(block-d-j) for j in range(int((block-d-payload)/2+1), block-d)])
            val = nCr(block, d) * rsdrop**d * corrupt
            rsfail.append(val)
        prfail.append(sum(rsfail))
    reedfail = filter(lambda x: x < param, prfail)
    maxpd.append(pd[(len(reedfail)-1)])

# for k in xrange(0, 237):
#     reeddrop = [sum([nCr(20+k, d)* hcf**d *(1-hcf)**(20+k-d) for d in range(int(k/2)+1, 20+k)]) for hcf in hammfail]
#     reedfail = filter(lambda x: x < param, reeddrop)
#     maxpd.append(pd[(len(reedfail)-1)])
    
# figure()
stem(rates, maxpd, lw=2.0)
xlabel('Rate', fontsize=20)
ylabel('Max Bit Error Probability', fontsize=18, labelpad=15)
title('Overall Tolerable Drop Pr = {0}'.format(param), fontsize=24)

comparison.append((rates, maxpd))
```
