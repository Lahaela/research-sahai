```python
def optimization(num_nodes, SNR_range):
    #downtarget, uptarget = logsearch(10**(-8), 10**(-10), 20)
    downtarget, uptarget = 1 * 10**(-9), 1 * 10**(-9)
    downFile = '../data_tables/optimize_down/n' + str(num_nodes) + '.in'
    downNode = load_table(downFile)
    upFile = '../data_tables/optimize_up/n' + str(num_nodes) + '.in'
    upNode = load_table(upFile)
    bitrange = sorted(downNode.tables.keys())
    upbitrange = sorted(upNode.tables.keys())
    xorbitlst = []
    for SNR in SNR_range:
        downbit, upbit = float("inf"), float("inf")
#         print(bitrange)
        for bit in bitrange:
            bittable = downNode.tables[bit]
            func = interp1d(bittable[0], bittable[1], kind='linear', bounds_error=False, fill_value=(1.0, 0.0))
            if func(SNR) <= downtarget:
                downbit = bit
                break
#         print('SNR', SNR, 'downbit', downbit)
        for bit in upbitrange:
            bittable = upNode.tables[bit]
            func = interp1d(bittable[0], bittable[1], kind='linear', bounds_error=False, fill_value=(1.0, 0.0))
            if func(SNR) <= uptarget:
                upbit = bit
                break
        xorbit = max(0, 40000 - downbit - upbit)
        blocklength = int(xorbit / 21)
        rate = 1 * 160 / xorbit * 7 / 4 if xorbit else float("inf")
        if rate > 1: continue
        k = (1-rate)*blocklength
        pbitdrop = Q(np.sqrt(2*10**(SNR/10)))
        hcerr = 1 - ((1-pbitdrop)**7 + 7*pbitdrop*(1-pbitdrop)**6)
        hcf = 1 - (1-hcerr)**3
        reeddrop = sum([nCr(blocklength, d)* hcf**d *(1-hcf)**(blocklength-d) for d in range(int(k/2)+1, blocklength)])
        if reeddrop <= 10**(-9):
            return array([SNR, downbit, upbit, xorbit])
#         xorbitlst.append(xorbit)
#     return xorbitlst
        # do xor analysis
        # if func(SNR) <= target: save bit1, bit2, alpha, beta
```

```python
# Final

p_add_1 = 1 * 10**(-9)
p_add_2 = 1 * 10**(-9)
p_add_3 = 0.1 * 10**(-9)

def xor_analysis(SNR, N):
    downfile = '../data_tables/hs_rs_downlink/n' + str(N) + '.in'
    downtable = load_table(downfile)
    SNR_down = downtable[0][np.where(np.array(downtable[1])<=p_add_1)[0][0]]
    
    upfile = '../data_tables/hs_rs_uplink/n' + str(N) + '.in'
    uptable = load_table(upfile)
    SNR_up = uptable[0][np.where(np.array(uptable[1])<=p_add_2)[0][0]]
    
    hcrit_1 = 10**((SNR_down - SNR)/10) # linear fade
    p_fade_1 = 1 - np.exp(-hcrit_1)
    hcrit_2 = 10**((SNR_up - SNR)/10) # linear fade
    p_fade_2 = 1 - np.exp(-hcrit_2)
    p_link_1 = p_fade_1 + (1 - p_fade_1) * p_add_1
    p_link_2 = p_fade_2 + (1 - p_fade_2) * p_add_2
    q = p_add_1 / p_link_1
    
    result = 0
    for ad in range(1, N+1):
        p_ad = nCr(N, ad) * (1 - p_link_1)**ad * p_link_1**(N-ad)
        adn_result = 0
        for adu in range(1, ad+1):
            p_adu = nCr(ad, adu) * (1 - p_add_2)**adu * p_add_2**(ad-adu)
            p_adn_success = (1 - p_add_3)**(ad - adu)
            adn_result += p_adu * p_adn_success
            
        b_result = 0
        for b in range(0, N-ad+1):
            bu_result = 0
            for bu in range(0, b+1):
                p_bu = nCr(b, bu) * (1 - p_add_2)**bu * p_add_2**(b-bu)
                p_bu_success = (1 - p_add_3)**bu
                p_rest_success = ((1 - p_link_2**ad) * (1 - p_add_3)**2)**(N-ad-bu)
                bu_result += p_bu * p_bu_success * p_rest_success
                
            p_b = nCr(N-ad, b) * q**b * (1-q)**(N-ad-b)
            b_result += p_b*bu_result
            
        result += p_ad * adn_result * b_result
    return result
```

```python
%%time
start_SNR = 0
target = 10**(-9)
dSNR = 0.01
p_add_1 = 10**(-3)
p_add_2 = 10**(-10)
N = 32
downfile = '../data_tables/optimize_down/n' + str(N) + '.in'
downtable = load_table(downfile)
down = downtable.tables
bitrange = sorted(down.keys())[1:]

snr_opt = []
for b in bitrange:
    snr_1 = down[b][0][np.where(np.array(down[b][1])<=p_add_1)[0][0]]
    snr_2 = down[20000-b][0][np.where(np.array(down[20000-b][1])<=p_add_2)[0][0]]
#     print snr_1, snr_2
    
    SNR = start_SNR
    pprotocol = 1.0
    while pprotocol > target:
        SNR += dSNR
        hcrit_1 = 10**((snr_1 - SNR)/10) # linear fade
        hcrit_2 = 10**((snr_2 - SNR)/10)
        pbadfade_1 = 1 - np.exp(-hcrit_1)
        pbadfade_2 = 1 - np.exp(-hcrit_2)
        psingle = pbadfade_1 + (1-pbadfade_1)*p_add_1

        pprotocol = sum([nCr(N, a) * (1-psingle)**a * psingle**(N-a) * 
                             (1-(1-(pbadfade_2**a + (1-pbadfade_2**a) * p_add_2))**(N-a)) for a in range(N)])
    snr_opt.append(SNR)
#     print hcrit_1, hcrit_2
#     print psingle
#     print pbadfade_1, pbadfade_2
#     print pprotocol
    
```
